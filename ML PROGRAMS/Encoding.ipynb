Encoding Techniques

Label Encoding

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
```

```python
df=pd.read_csv("satisfaction_dataset.csv")
df.head(10)
```

```python
LE=LabelEncoder()
LE.fit(df.iloc[:,1])
```

```python
df.iloc[:,1]=LE.transform(df.iloc[:,1])
df.drop("Satisfaction_Encoded",axis=1)
```

Ordinal Encoding

```python
df1=pd.read_csv("satisfaction_dataset.csv")
```

```python
OE = OrdinalEncoder(categories= [['Poor', 'Average', 'Good', 'Excellent']])
```

```python
OE.fit(df1.iloc[:,1].values.reshape(-1,1))
df1.iloc[:,1]=OE.transform(df1.iloc[:,1].values.reshape(-1,1))
print(df1)
```
