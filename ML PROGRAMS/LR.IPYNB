LINEAR REGRESSION
       

Predicting house prices


IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

READING DATASET

```python
data=pd.read_csv("houseprices.csv")
data.head()
```

CREATING MODEL

```python
model=LinearRegression()
model.fit(data[["area"]],data.price)
```

PREDICTION FOR ONE VALUE

```python
value=3500
x=np.array(value).reshape(1,-1)
model.predict(x)
```

```python
X_line = np.linspace(min(data['area']), max(data['area']), 100).reshape(-1, 1)
y_line = model.predict(X_line)
```

VISUALIZING THE OUTPUT

```python
x=data[["area"]]
y=data[["price"]]
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(X_line, y_line, color='red', label='Regression Line')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.title('Linear Regression: Area vs Price')
plt.legend()
plt.grid(True)
plt.show()
```

PREDICTION FOR MULTIPLE VALUES

```python
df=pd.read_csv("areas.csv")
df.head()
```

```python
price= model.predict(df[["area"]])
df["prices"]=price
df
```
